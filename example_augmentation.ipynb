```python
import pandas as pd
import os

# List of files
files = ["augmented_data.csv",
    "augmented_data2.csv", "augmented_data3.csv",
    "augmented_data4.csv", "augmented_data5.csv",
    "augmented_data7.csv", "augmented_data6.csv",
    "augmented_data8.csv", "augmented_data9.csv",
    "augmented_data10.csv", "augmented_data11.csv","augmented_data12.csv", "augmented_data14.csv",
    "augmented_data15.csv", "augmented_data16.csv"
]

# Function to load files
def load_files(file_list):
    dataframes = {}
    for file in file_list:
        file_path = os.path.join(file)
        if file.endswith('.csv'):
            # Read CSV into DataFrame
            dataframes[file] = pd.read_csv(file_path)
        elif file.endswith('.json'):
            # Read JSON into DataFrame
            print(file)
            dataframes[file] = pd.read_json(file_path)
        elif file.endswith('.jsonl'):
            # Read JSONL into DataFrame
            dataframes[file] = pd.read_json(file_path, lines=True)
    return dataframes

# Load all files
dataframes = load_files(files)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_class_distribution(df, col):
    # Count the number of records for each unique subcategory
    subcategory_counts = df[col].value_counts()
    print(subcategory_counts)
```

```python
# Display loaded dataframes
augmented_dfs=[]
for name, df in dataframes.items():
    print("******************************************************************************")
    print(f"DataFrame for {name}:")
    if "subcategory" in df.columns:
      plot_class_distribution(df,"subcategory")
      augmented_dfs.append(df)
    if "category" in df.columns:
      plot_class_distribution(df,"category")
    if "labels" in df.columns:
      plot_class_distribution(df,"labels")
```

```python
import json

def load_data(train=True):
    if train:
        name_data = "training.json"
    else:
        name_data = "test.json"

    # Read the data from the specified JSON file
    with open(name_data, "r", encoding='utf-8') as json_file:
        data = json.load(json_file)
    return data

train_data=load_data()
test_data=load_data(train=False)

def convert_to_pandas(data):
  records = []
  for item in data:
      text = item['text']
      category = item['labels'][0]['category']
      subcategories = ', '.join(item['labels'][0]['subcategories'])  # Join subcategories into a single string
      records.append({'text': text, 'category': category, 'subcategories': subcategories})

  return pd.DataFrame(records)

train_df=convert_to_pandas(train_data)
test_df=convert_to_pandas(test_data)
```

```python
concat_df = pd.concat([train_df,test_df], ignore_index=True)
concat_df
```

```python
plot_class_distribution(concat_df,"subcategories")
```
