```python
# Function to find the category and subcategory of a given subcategory name
import unicodedata
import os

def normalize_string(s):
    # Normalize unicode characters to NFC form
    s = unicodedata.normalize("NFC", s)
    return s.strip()

def find_category_and_subcategory(subcategory_name):
    normalized_name = normalize_string(subcategory_name)
    print(f"Searching for normalized name: '{normalized_name}'")

    for category in categories:
        for subcategory in category["subcategories"]:
            print(f"Comparing '{normalized_name}' with '{subcategory}' (lengths: {len(normalized_name)}, {len(subcategory)})")
            if normalized_name == subcategory:
                print(f"FOUND {normalized_name}")
                return {"category": category["category"], "subcategories": [subcategory]}

    print(f"not found {normalized_name}")
    return None

def save_data(data,train=True):
    if train :
      name_data="training.json"
    else :
      name_data="test.json"
    # Save the training data in JSON format
    with open(name_data, "w", encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

    print(f"data saved to {name_data}")

def load_data(train=True):
    if train:
        name_data = "training.json"
    else:
        name_data = "test.json"

    # Read the data from the specified JSON file
    with open(name_data, "r", encoding='utf-8') as json_file:
        data = json.load(json_file)
    return data
def construct_data(train=True):
    # Prepare the list to hold the training data
    training_data = []

    # Traverse the "training" folder
    if train:
      data_dir = "training"
    else:
      data_dir="test"
    for subcategory_folder in os.listdir(data_dir):
        subcategory_path = os.path.join(data_dir, subcategory_folder)

        label = find_category_and_subcategory(subcategory_folder)
        print(label)
        if label:  # Only proceed if the subcategory is in our defined categories
            for filename in os.listdir(subcategory_path):
                if filename.endswith(".txt"):
                    file_path = os.path.join(subcategory_path, filename)
                    with open(file_path, 'r', encoding='utf-8') as file:
                        text_content = file.read().strip()

                        # Create the structured data entry
                        entry = {
                            "text": text_content,
                            "labels": [label]
                        }
                        training_data.append(entry)
    save_data(training_data,train)
```

```python
import pandas as pd
import random

def balance_subcategories(encoded_df, max_records_per_subcategory=700):
    """
    Balances the dataset by keeping a maximum of `max_records_per_subcategory` records per subcategory.

    Parameters:
    - encoded_df: Pandas DataFrame containing the encoded data.
    - max_records_per_subcategory: Maximum number of records to keep per subcategory.

    Returns:
    - balanced_df: A new DataFrame with balanced subcategories.
    """
    # Group the DataFrame by 'subcategory_label'
    grouped = encoded_df.groupby('subcategory_label')

    # Initialize an empty list to store the balanced data
    balanced_data = []

    # Iterate over each subcategory
    for subcategory, group in grouped:
        # If the group has more than `max_records_per_subcategory` records, sample randomly
        if len(group) > max_records_per_subcategory:
            sampled_group = group.sample(n=max_records_per_subcategory, random_state=42)
        else:
            # Otherwise, keep all records in the group
            sampled_group = group

        # Append the sampled group to the balanced data list
        balanced_data.append(sampled_group)

    # Concatenate all the sampled groups into a single DataFrame
    balanced_df = pd.concat(balanced_data, ignore_index=True)

    return balanced_df
    # Assuming `encoded_df` is your DataFrame
balanced_df = balance_subcategories(encoded_df)

# Print the shape of the balanced DataFrame
print(f"Original DataFrame shape: {encoded_df.shape}")
print(f"Balanced DataFrame shape: {balanced_df.shape}")

# Check the distribution of subcategories in the balanced DataFrame
print(balanced_df['subcategory_label'].value_counts())
```

```python
import torch
from torch.utils.data import Dataset
from transformers import AutoTokenizer, AutoModel
from sklearn.model_selection import train_test_split

MAX_LENGTH=256

class TextEmbeddingDataset(Dataset):
    def __init__(self, data, max_length=MAX_LENGTH):
        """
        Initialize the dataset.

        Parameters:
        - data: DataFrame containing encoded text and labels.
        - max_length: Maximum sequence length for padding and truncation.
        """
        self.data = data
        self.model = arabert_model

    def __len__(self):
        """Return the length of the dataset."""
        return len(self.data)

    def __getitem__(self, idx):
        """
        Get the embeddings and labels for a specific index.

        Parameters:
        - idx: Index of the data point in the DataFrame.

        Returns:
        - A dictionary with embeddings and labels.
        """
        # Extract encoded tokens from the DataFrame
        input_ids = torch.tensor(self.data.iloc[idx]['input_ids']).unsqueeze(0)
        attention_mask = torch.tensor(self.data.iloc[idx]['attention_mask']).unsqueeze(0)
        token_type_ids = torch.tensor(self.data.iloc[idx]['token_type_ids']).unsqueeze(0)

        # Get embeddings from the AraBERT model
        with torch.no_grad():
            outputs = self.model(input_ids=input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids)

        # Use the embedding of the [CLS] token as the representation of the text
        embeddings = outputs.last_hidden_state[:, 0, :]  # Shape: [1, embedding_dim]

        # Retrieve the category and subcategory labels
        category_label = self.data.iloc[idx]['category_label']
        subcategory_label = self.data.iloc[idx]['subcategory_label']

        return {
            'embeddings': embeddings.squeeze(0),  # Remove batch dimension
            'category_label': torch.tensor(category_label, dtype=torch.long),
            'subcategory_label': torch.tensor(subcategory_label, dtype=torch.long)
        }


# Split dataset into training and validation sets
train_data, val_data = train_test_split(encoded_df, test_size=0.2, random_state=42)

# Create datasets
train_dataset = TextEmbeddingDataset(train_data)
val_dataset = TextEmbeddingDataset(val_data)
```

```python
import random

class TextClassificationEnv:
    def __init__(self, dataset, model, categories, subcategories, max_steps=2):
        """
        Initialize the environment.

        Parameters:
        - dataset: PyTorch dataset containing embeddings and labels.
        - model: The model used for predictions (although not directly used in this class).
        - categories: List of possible main categories.
        - subcategories: Dictionary of subcategories per category.
        - max_steps: Max steps per episode (for hierarchical classification).
        """
        self.dataset = dataset
        self.model = model
        self.categories = categories
        self.subcategories = subcategories
        self.max_steps = max_steps
        self.reset()

    def reset(self):
        """Resets the environment for a new episode."""
        # Randomly select a new record from the dataset
        self.current_index = random.randint(0, len(self.dataset) - 1)

        # Extract embeddings and labels from the dataset
        sample = self.dataset[self.current_index]
        self.current_embeddings = sample['embeddings']

        # Set true labels for category and subcategory
        self.true_category = sample['category_label'].item()
        self.true_subcategory = sample['subcategory_label'].item()

        # Reset the step counter
        self.steps = 0

        # Return initial observation (embedding vector)
        return self.current_embeddings

    def step(self, action, is_category_step):
        done = False
        reward = 0

        if is_category_step:
            # Category prediction
            predicted_category = action
            self.predicted_category = predicted_category
            if predicted_category == self.true_category:
                reward += 2  # Correct category
            else:
                reward -= 1  # Incorrect category
        else:
            # Subcategory prediction
            predicted_subcategory = action
            if predicted_subcategory == self.true_subcategory:
                reward += 1.5
            elif self.true_category == self.predicted_category:
                reward += 0.5  # Partial correctness
            else:
                reward -= 0.5
            done = True

        self.steps += 1
        if self.steps >= self.max_steps:
            done = True

        # Add step type to the state
        new_state = {
            'embeddings': self.current_embeddings,
            'step_type': 0 if is_category_step else 1,
            'predicted_category': getattr(self, 'predicted_category', -1)
        }
        return new_state, reward, done

    def sample_action(self):
        if self.steps == 0:
            # Sample category
            return random.choice(self.categories)
        elif hasattr(self, 'predicted_category') and self.predicted_category in self.subcategories:
            # Sample subcategory for predicted category
            return random.choice(list(self.subcategories[self.predicted_category].keys()))
        else:
            # Fallback to general subcategories
            category = random.choice(self.categories)
            return random.choice(list(self.subcategories[category].keys()))
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class CategoryNetwork(nn.Module):
    def __init__(self, input_dim=EMBEDDING_DIM, num_categories=5, hidden_dim=256, num_heads=4):
        super(CategoryNetwork, self).__init__()
        self.attention = nn.MultiheadAttention(embed_dim=input_dim, num_heads=num_heads, batch_first=True)
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.output = nn.Linear(hidden_dim // 2, num_categories)

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m, nn.MultiheadAttention):
                nn.init.xavier_uniform_(m.in_proj_weight)
                nn.init.zeros_(m.in_proj_bias)
                nn.init.xavier_uniform_(m.out_proj.weight)
                nn.init.zeros_(m.out_proj.bias)

    def forward(self, x):
        """
        Forward pass for category prediction.

        Parameters:
        - x: Input embedding (1D tensor).

        Returns:
        - Q-values for all categories (1D tensor).
        """
        # Normalize input to prevent large magnitudes
        x = F.layer_norm(x, x.size())

        # Add sequence dimension for attention (seq_len=1)
        x = x.unsqueeze(0)  # Shape becomes (seq_len=1, embedding_dim)

        # Self-attention layer
        attn_output, _ = self.attention(x, x, x)  # Self-attention with single element
        x = attn_output.squeeze(0)  # Remove sequence dimension

        # Fully connected layers
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.output(x)  # Q-values for categories


class SubcategoryNetwork(nn.Module):
    def __init__(self, input_dim=EMBEDDING_DIM, num_subcategories=subcategory_counts, hidden_dim=256, num_heads=4):
        super(SubcategoryNetwork, self).__init__()
        self.attention = nn.MultiheadAttention(embed_dim=input_dim, num_heads=num_heads, batch_first=True)
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, hidden_dim // 2)
        self.output = nn.Linear(hidden_dim // 2, max(num_subcategories))  # Handle max subcategories dynamically
        self.num_subcategories = num_subcategories

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m, nn.MultiheadAttention):
                nn.init.xavier_uniform_(m.in_proj_weight)
                nn.init.zeros_(m.in_proj_bias)
                nn.init.xavier_uniform_(m.out_proj.weight)
                nn.init.zeros_(m.out_proj.bias)

    def forward(self, x, category_idx):
        """
        Forward pass for subcategory prediction.

        Parameters:
        - x: Input embedding (1D tensor).
        - category_idx: Index of the selected category.

        Returns:
        - Masked Q-values for subcategories (1D tensor).
        """
        # Normalize input to prevent large magnitudes
        x = F.layer_norm(x, x.size())

        # Add sequence dimension for attention (seq_len=1)
        x = x.unsqueeze(0)  # Shape becomes (seq_len=1, embedding_dim)

        # Self-attention layer
        attn_output, _ = self.attention(x, x, x)  # Self-attention with single element
        x = attn_output.squeeze(0)  # Remove sequence dimension

        # Fully connected layers
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        subcategory_logits = self.output(x)

        # Mask irrelevant subcategories based on the selected category
        mask = torch.zeros_like(subcategory_logits)
        mask[:self.num_subcategories[category_idx]] = 1
        return subcategory_logits * mask
```

```python
def predict(state: torch.Tensor, next_state: torch.Tensor, action: int, reward: float, done: bool, subcategory_or_category: bool, selected_category : int):
    """
    Function to handle both category and subcategory prediction and update logic.
    The `subcategory_or_category` boolean determines which network to call.

    Parameters:
    - state (torch.Tensor): Current state embeddings.
    - next_state (torch.Tensor): Next state embeddings after taking the action.
    - action (int): Action taken by the agent.
    - reward (float): Reward received for taking the action.
    - done (bool): Whether the episode has ended.
    - subcategory_or_category (bool): If True, use the subcategory network; else use the category network.
    - selected_category (int, optional): Selected category for subcategory prediction.
    """

    # Choose network based on subcategory_or_category flag
    if subcategory_or_category:
        # Ensure selected_category is provided for subcategory prediction
        if selected_category is None:
            raise ValueError("Selected category must be provided for subcategory prediction.")

        # Get current Q-values for subcategory prediction
        current_q_values = subcategory_network(state, category_idx=selected_category)

        # Get next Q-values for subcategory prediction
        with torch.no_grad():
            next_q_values = subcategory_network(next_state, category_idx=selected_category)
    else:
        # Get current Q-values for category prediction
        current_q_values = category_network(state)

        # Get next Q-values for category prediction
        with torch.no_grad():
            next_q_values = category_network(next_state)

    # Ensure the action is within the bounds of the network's output
    if action >= current_q_values.shape[1]:
        padding = torch.zeros(1, action + 1 - current_q_values.shape[1], device=state.device)
        current_q_values = torch.cat((current_q_values, padding), dim=1)

    # Calculate max Q-value for the next state
    max_next_q_value = next_q_values.max().item()

    # Calculate the target Q-value using the Bellman equation
    target_q_value = reward + GAMMA* max_next_q_value * (1 - done)

    return current_q_values, target_q_value


def train_step(
    state: torch.Tensor,
    action: int,
    reward: float,
    next_state: torch.Tensor,
    done: bool,
    step: int,
    selected_category: int, optimizer = OPTIMIZER
):
    """
    Perform a single training step for Q-learning with category and subcategory Q-networks using a unified function.

    Parameters:
    - state (torch.Tensor): Current state embeddings.
    - action (int): Action taken by the agent.
    - reward (float): Reward received for taking the action.
    - next_state (torch.Tensor): Next state embeddings after the action.
    - done (bool): Whether the episode has ended.
    - step (int): Step of the process (0 for category, 1 for subcategory).
    - selected_category (int, optional): Selected category for subcategory prediction.
    """

    # Determine if we're predicting the category (step == 0) or subcategory (step == 1)
    subcategory_or_category = step == 1

    # Call the unified prediction function
    current_q_values, target_q_value = predict(state, next_state, action, reward, done, subcategory_or_category, selected_category)

    # Compute the loss: MSE between the current Q-value for the taken action and the target
    criterion = nn.SmoothL1Loss()(current_q_values[0, action], torch.tensor([target_q_value], device=state.device))

    # Backpropagation and optimizer step
    optimizer.zero_grad()
    criterion.backward()
    optimizer.step()

    return criterion.item()  # Return loss for monitori
```

```python
def train_q_network(category_network, subcategory_network, optimizer=OPTIMIZER, env=env, num_episodes=9000):
    """
    Train the Q-network using a reinforcement learning approach with epsilon-greedy exploration.
    """
    # Initialize epsilon parameters
    # For slower convergence to exploitation
    INITIAL_EPSILON = 1.0       # Start with 100% exploration
    FINAL_EPSILON = 0.05       # Keep minimum exploration at 5%
    EPSILON_DECAY = 0.999  # Very slow decay rate
    current_epsilon = INITIAL_EPSILON

    episode_rewards = []
    losses = []
    val_rewards = []

    # Set networks to training mode
    category_network.train()
    subcategory_network.train()

    for episode in range(num_episodes):
        state = env.reset()
        done = False
        episode_reward = 0
        steps = 0
        selected_category = None

        print(f"--- Episode {episode + 1} ---")
        print(f"Current Epsilon: {current_epsilon:.4f}")
        print(f"Initial State: tensor of shape {state.shape}")

        while not done:
            if isinstance(state, dict):
                state_tensor = state['embeddings'].unsqueeze(0).to(device)
            else:
                state_tensor = state.unsqueeze(0).to(device)

            # Epsilon-greedy action selection
            if np.random.rand() < current_epsilon:
                # Exploration: Random action
                if steps == 0:
                    action = env.sample_action()
                    selected_category = action
                    print(f"Exploring: Chose random category action {action}")
                else:
                    action = env.sample_action()
                    print(f"Exploring: Chose random subcategory action {action} for category {selected_category}")
            else:
                # Exploitation: Use the network for prediction
                with torch.no_grad():  # No need to track gradients during exploitation
                    if steps == 0:
                        q_values = category_network(state_tensor)
                        print(f"Category Q-values: {q_values}")
                        action = torch.argmax(q_values).item()
                        selected_category = action
                        print(f"Exploiting: Chose category action {action} with max Q-value")
                    else:
                        q_values = subcategory_network(state_tensor, selected_category)
                        print(f"Subcategory Q-values: {q_values}")
                        action = torch.argmax(q_values).item()
                        print(f"Exploiting: Chose subcategory action {action} with max Q-value")

            # Take action and observe next state
            next_state, reward, done = env.step(action, is_category_step=(steps == 0))
            next_state_tensor = torch.tensor(next_state['embeddings'], dtype=torch.float32).unsqueeze(0).to(device)
            print(f"Action taken: {action}, Reward: {reward}, Next state: {next_state['embeddings'].shape}, Done: {done}")

            # Store transition and train
            loss = train_step(state_tensor, action, reward, next_state_tensor, done,
                            step=steps, selected_category=selected_category, optimizer=optimizer)
            print("Loss:", loss)
            losses.append(loss)

            state = next_state
            episode_reward += reward
            steps += 1

        # Decay epsilon after each episode
        current_epsilon = max(FINAL_EPSILON, current_epsilon * EPSILON_DECAY)

        episode_rewards.append(episode_reward)
        print(f"Episode {episode + 1} ended with Total Reward: {episode_reward}")
        print(f"New Epsilon Value: {current_epsilon:.4f}\n")

    # Plotting and analysis
    plt.figure(figsize=(18, 9))
    plt.subplot(2, 1, 1)
    plt.plot(episode_rewards, label="Episode Reward")
    plt.xlabel("Episode")
    plt.ylabel("Total Reward")
    plt.title("Learning Curve")
    plt.legend()
    plt.grid()

    plt.subplot(2, 1, 2)
    plt.plot(losses, label="Loss", color='red')
    plt.xlabel("Episode")
    plt.ylabel("Loss")
    plt.title("Training Loss")
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

    # Calculate final metrics
    positive_rewards = sum(1 for r in episode_rewards if r > 0)
    total_rewards = len(episode_rewards)
    reward_accuracy = positive_rewards / total_rewards if total_rewards > 0 else 0
    print(f"Final Epsilon: {current_epsilon:.4f}")
    print(f"Cumulative Reward Accuracy: {reward_accuracy:.2f} ({positive_rewards}/{total_rewards})")

    # Save trained models
    save_trained_networks(category_network, subcategory_network)

    # Set networks to evaluation mode after training
    category_network.eval()
    subcategory_network.eval()

def save_trained_networks(category_network, subcategory_network):
    """Save the trained networks to disk."""
    torch.save(category_network.state_dict(), 'category_network.pth')
    torch.save(subcategory_network.state_dict(), 'subcategory_network.pth')
    print("Trained networks saved as 'category_network2.pth' and 'subcategory_network2.pth'.")
```

```python
import torch

def classify_text(state, category_network, subcategory_network):
    """
    Classify a text input into a high-level category (label) and subcategory (sub_label).

    Parameters:
    - category_network: The trained neural network for category classification.
    - subcategory_network: The trained neural network for subcategory classification.

    Returns:
    A dictionary with the predicted label and sub_label.
    """
    # Step 2: Predict the high-level category (label) using the category network
    category_q_values = category_network(state)  # This should output a tensor of Q-values for the categories
    category_action = torch.argmax(category_q_values, dim=-1).item()  # Choose the category with the highest Q-value

    # Step 3: Predict the subcategory (sub_label) using the subcategory network
    subcategory_q_values = subcategory_network(state, category_action)  # Pass the category action to the subcategory network
    subcategory_action = torch.argmax(subcategory_q_values, dim=-1).item()  # Choose the subcategory with the highest Q-value

    # Step 4: Return the results
    return {"label": category_action, "sub_label": subcategory_action}
```
